# Carga de datos con open

```python
data3 = open(filename, "r")
```

```python
cols = data3.readline().strip().split(",")
```

```python
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
    
print("el data set %d filas y %d columnas" % (counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

## Lectura y escritura

```python
infile = "Customer Churn Model.txt"
outfile = "Tab Customer Churn Model.txt"
```

```python
with open(infile, "r") as infilel:
    with open(outfile, "w") as outfilel:
        for line in infilel:
            fields = line.strip().split(",")
            outfilel.write("\t".join(fields))
            outfilel.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep = "\t")
df4.head()
```
